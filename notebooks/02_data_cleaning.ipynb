## Data Cleaning

### 1. Project Objective

The purpose of this notebook is to improve data quality and prepare the dataset for exploratory data analysis.

The cleaning process includes:

- Assessing data quality
- Correcting data types
- Creating useful features
- Exporting a cleaned dataset

### 2. Import Libraries

```python
import pandas as pd
import numpy as np
```

### 3. Load Raw Dataset

```python
df = pd.read_csv("../data/raw/Superstore.csv", encoding="ISO-8859-1")
df.head(5)
```

### 4. Assess Data Quality

#### 4.1 Missing Values

```python
missing = df.isna().sum()
missing
```

```python
missing[missing > 0]
```

No missing values were found.

#### 4.2 Duplicate Records

```python
df.duplicated().sum()
```

No duplicate records were found.

#### 4.3 Data Types

```python
df.info()
```

Order Date and Ship Date are stored as object and need to be converted to datetime.

#### 4.4 Numeric Columns

```python
numeric_columns = [
    "Sales",
    "Quantity",
    "Discount",
    "Profit"
]

df[numeric_columns].dtypes
```

### 5. Clean Data

#### 5.1 Convert Date Columns

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format = '%d-%m-%Y'
)

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    format = '%d-%m-%Y'
)
```

```python
df.info()
```

#### 5.2 Standardize Data Types

```python
df["Postal Code"] = df["Postal Code"].astype("string")
```

```python
df["Postal Code"].dtype
```

#### 5.3 Create New Features

```python
# Year
df["Year"] = df["Order Date"].dt.year

# Month
df["Month"] = df["Order Date"].dt.month

# Month Name
df["Month Name"] = df["Order Date"].dt.month_name()

# Quarter
df["Quarter"] = df["Order Date"].dt.quarter

# Weekday
df["Weekday"] = df["Order Date"].dt.day_name()
```

```python
# Shipping Days
df["Shipping Days"] = (
    df["Ship Date"]
    - df["Order Date"]
).dt.days
```

```python
# Profit Margin
df["Profit Margin"] = np.where(
    df["Sales"] != 0,
    df["Profit"] / df["Sales"],
    np.nan
)
```

```python
df.head(5)
```

### 6. Validate Cleaned Data

```python
df.info()
```

```python
df.describe().T
```

### 7. Export Clean Dataset

```python
df.to_csv(
    "../data/processed/superstore_clean.csv",
    index=False
)
```

The cleaned dataset has been exported successfully and will be used for subsequent exploratory data analysis.
